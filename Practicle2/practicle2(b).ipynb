```python
dataset1['species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2},inplace=True)
```

```python
dataset1 = dataset1.drop('species',axis='columns')

dataset1
```

```python
dataset1['sepal_length']= dataset1['sepal_length'].astype(float)
dataset1['sepal_width']= dataset1['sepal_width'].astype(float)
dataset1['petal_length']= dataset1['petal_length'].astype(float)
dataset1['petal_width']= dataset1['petal_width'].astype(float)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(dataset1,target,test_size=0.3)
```

```python

class Perceptron:
  def __init__(self,lr_rate,epochs):
    self.weights=None
    self.bias=None
    self.lr_rate = lr_rate
    self.epochs = epochs

  def activation(self,z):
      return np.heaviside(z,0)

  def fit(self,X,y):
      n_samples,n_features = X.shape
      self.weights = np.zeros(n_features)
      self.bias = 0

      for epoch in range (self.epochs):
          for i in range(n_samples):
              # z=np.dot(X.iloc[i], self.weights)+self.bias
              # y_pred = self.activation(z)
              z = np.dot(X.iloc[i], self.weights) + self.bias
              y_pred = self.activation(z)

              #updating weights and bias
              self.weights += self.lr_rate*(y.iloc[i]-y_pred)*X.iloc[i]
              self.bias += self.lr_rate*(y.iloc[i]-y_pred)

  def predict(self,X):
      z = np.dot(X,self.weights +self.bias)
      return self.activation(z)
```

```python
perceptron = Perceptron(0.01,15)
perceptron.fit(x_train,y_train)
```

```python
pred =perceptron.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,pred)
from sklearn.metrics import classification_report
print(classification_report(y_test,pred))
```
